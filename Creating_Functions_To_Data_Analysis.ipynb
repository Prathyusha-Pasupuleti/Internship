```python
def matchesPlayedByTown1():
  townName = []
  matchesPlayed = []
  TownData = df['home_team'].value_counts()
  for _ in TownData.index:
    if TownData[_] > 30:
     townName.append(_)
     matchesPlayed.append(TownData[_])
  plt.bar(matchesPlayed,townName)
  plt.ylabel('Town Name')
  plt.xlabel('Matches Played')
  plt.title('Matches Played by Town1')
  plt.show()
```

```python
#Function to get graph
def getGraph(DataFrame, Column1, Column2):
  from matplotlib import pyplot as plt

  #Check the length of Column1 and Column2
  if len(DataFrame[Column1]) == len(DataFrame [Column2]):
    df.plot(x =Column1,y=Column2,kind = 'bar',color=['green','orange'])
    print('The graph/plot is successful created ')
  else:
    print('Please check the column size once')
```
