```python
def AnalyzeFrance():
    matchesPlayed = df['home_team'].value_counts()['France']
    print(f'total Number of matchs played by France is : {matchesPlayed}')
```

```python
def AnalyzeTown():
  TownData = df['home_team'].value_counts()
  for _ in TownData.index:
    print(f'Total Number of matchs played by (_) is : {TownData[_]}')
```

```python
AnalyzeTown()
```

```python
townName = []
matchesPlayed = []
TownData = df['home_team'].value_counts()
for _ in TownData.index:
  if TownData[_] > 30:
    townName.append(_)
    matchesPlayed.append(TownData[_])
```

```python
townName
```

```python
from matplotlib import pyplot as plt
```

```python
plt.bar(matchesPlayed,townName)
plt.ylabel('Town Name')
plt.xlabel('Matches Played')
plt.title('Matches Played by Town')
plt.show()
```

```python
def matchesPlayedByTown1():
  townName = []
  matchesPlayed = []
  TownData = df['home_team'].value_counts()
  for _ in TownData.index:
    if TownData[_] > 30:
     townName.append(_)
     matchesPlayed.append(TownData[_])
  plt.bar(matchesPlayed,townName)
  plt.ylabel('Town Name')
  plt.xlabel('Matches Played')
  plt.title('Matches Played by Town1')
  plt.show()

```

```python
matchesPlayedByTown1()
```

```python
def matchesPlayedByTown2():
  Above30 = df['home_team'].value_counts()
  filteredData = Above30[Above30 > 30]
  filteredData.plot(kind ='barh',title = 'Matches Played by Town2')
```
